```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from warnings import filterwarnings
sns.set_theme()
%config InlineBackend.figure_format = 'retina'
filterwarnings('ignore')
```

```python
train=pd.read_csv('../input/30-days-of-ml/train.csv')
test=pd.read_csv('../input/30-days-of-ml/test.csv')
```

```python
x_train=train.drop(['target'],axis=1)
```

```python
x_train.head()
```

```python
y_train=train['target']
```

```python
x_test=test.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
categorical_cols=train.select_dtypes(include=['object']).columns.values
numerical_cols=train.select_dtypes(include=['float64']).columns.values
```

```python
for i in categorical_cols:
    x_train[i]=label.fit_transform(x_train[i])
    
for i in categorical_cols:
    x_test[i]=label.fit_transform(x_test[i])
```

```python
x_train.head()
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_valid,y_train1,y_valid=train_test_split(x_train,y_train,test_size=0.2,random_state=1)
```

```python
from xgboost import XGBRFRegressor,XGBRegressor
model=XGBRegressor(base_score=0.2,gamma=0.2,learning_rate=0.07,max_depth=3,reg_alpha=21,n_estimators=2000,n_jobs=-1,random_state=1,tree_method='gpu_hist')
model.fit(x_train1,y_train1)
```

```python
pred=model.predict(x_valid)
pred
```

```python
pred.shape
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('Mean Squared error =', mean_squared_error(y_valid,pred,squared=False))
```

```python
# Use the model to generate predictions
predictions = model.predict(x_test)

# Save the predictions to a CSV file
output = pd.DataFrame({'Id': x_test.id,
                       'target': predictions})
output.to_csv('submission.csv', index=False)
```
